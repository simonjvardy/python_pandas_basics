### Pandas read_csv() with no Header Row

```python
df6=pandas.read_csv('supermarkets-commas_no_header.txt')
df6
```

***As can be seen above, the first data row has been used as the header row which is not correct. To overcome this error, use attribute `header=None`.***

```python
df7=pandas.read_csv('supermarkets-commas_no_header.txt', header=None)
df7
```

***Set the Column Names on the data frame***

```python
df7.columns = ['ID', 'Address','City','State','Country','Name','Employees']
df7
```

### Set the Index Column ###

```python
df7.set_index("ID")
```

```python
df7
```

***Using `set_index(col_name)` is only performed 'on-the-fly'. Returning df7 again shows the original data frame contents.
To permanently store the index, either assign to a new data frame or use the `inplace=True` attribute...***

```python
df8 = df7.set_index("ID")
df8
```

```python
df7.set_index("ID", inplace=True)
df7
```

### Filtering Data from a Pandas Dataframe

***Index based filtering***

- Row index 3 to 4 and columns Name and Employee

```python
df7.loc[3:4,"Name":"Employees"]
```

- All rows for the listed columns

```python
df7.loc[:,"Name":"Employees"]
```

- Create a Python List of all rows in the Employee column

```python
list(df7.loc[:,"Employees"])
```
